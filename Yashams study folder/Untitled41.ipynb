### Method 1: Using `IPython.display.Image`

This method is useful for displaying images directly within the notebook output, especially if you have an image file or a URL.

```python

import requests
from IPython.display import Image

image_url = 'https://www.google.com/images/branding/googlelogo/1x/googlelogo_color_272x92dp.png'
image_path = 'google_logo.png'

response = requests.get(image_url)
with open(image_path, 'wb') as f:
    f.write(response.content)

print(f"Displaying image from {image_path}:")
Image(filename=image_path)
```

### Method 2: Using `matplotlib.pyplot.imshow`

This method is great for more complex image manipulations and visualizations, often used with `PIL` (Pillow) for image loading.

```python
import matplotlib.pyplot as plt
from PIL import Image as PIL_Image

img = PIL_Image.open('google_logo.png')

plt.imshow(img)
plt.axis('off')
plt.show()
```
