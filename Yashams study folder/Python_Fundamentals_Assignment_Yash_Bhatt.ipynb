###1. Write a program that checks if a number is positive. If it is, print 'Positive Number'

```python
number = 5
if number > 0:
    print("Positive Number")
```

###2. Create an if - else statement that checks if a person is eligible to vote (vote >= 18)

```python
age = 15
if age >= 18:
    print("You are eligible to vote")
else:
    print("You are not eligible to vote")
```

###3. Write an if - elif - else condition to categorize a number as positive, negative, or zero

```python
number = 7
if number > 0:
    print("Positive")
elif number < 0:
    print("Negative")
else:
    print("Zero")
```

###4. Create a program that checks if a number is even or odd using if - else

```python
number = 8
if number % 2 == 0:
    print("Even")
else:
    print("Odd")
```

###5. Use a nested if condition to check if a number is positive & even

```python
number = 21
if number > 0:
    if number % 2 == 0:
        print("Positive and Even")
    else:
        print("Positive but Odd")
else:
    print("Not Positive")
```

###6. Write a one - line if condition to print 'python is cool' if x = 10

###7. Implement a one - line if - else condition to print 'Even' if a number is even, otherwise 'Odd'

```python
number = 27
print("Even" if number % 2 == 0 else "Odd")
```

###8. Write an if statement that checks if a given string contains the word 'Python'

###9. Check whether a number is divisible by both 3 & 5 using if - else

```python
number = 45
if number % 3 == 0 and number % 5 == 0:
    print("Divisible by both 3 and 5")
else:
    print("Not divisible by both 3 and 5")
```

###10. Write an if condition that checks whether a year is a leap year (divisible by 4 not 100, unless also divisible by 400)

```python
year = 2024
if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print(f"{year} is a leap year")
else:
    print(f"{year} is not a leap year")
```

###2. Create a program that takes two numbers and finds their remainder using the `%` operator.

###4. Write a program that checks if two variables have the same value using '==' operator

```python
print("\nProgram 4: Check equality using (==) operator")
x = 15
y = 15
is_equal = x == y
print(f"x = {x}, y = {y}")
print(f"x == y: {is_equal}")

```

###5. Implement a program that uses the `and` logical operator to check if both conditions `10 > 5` and `20 > 15` are `True`.

```python
print("\nProgram 5: Logical 'and' operator")
condition1 = 10 > 5
condition2 = 20 > 15
result = condition1 and condition2
print(f"(10 > 5) and (20 > 15): {result}")

```

###6. Use an assignment operator to increase the value of `x = 10` by `5` using `+=`

```python
print("\nProgram 6: Assignment operator (+=) ")
x = 10
print(f"Initial value of x: {x}")
x += 5
print(f"After x += 5: {x}")
```

###7. Write a program that checks if a number is **not** greater than `100` using the `not` operator

```python
print("\nProgram 7: Logical 'not' operator")
number = 50
result = not (number > 100)
print(f"number = {number}")
print(f"not (number > 100): {result}")
```

###8. Use `range()` and the `*` multiplication operator to print the first 10 multiples of `5`

```python
print("\nProgram 8: Multiplies of 5 using (*) operator")
print("First 10 multiples of 5:")
for i in range(1, 11):
    multiple = i * 5
    print(multiple, end=" ")
print()
```
