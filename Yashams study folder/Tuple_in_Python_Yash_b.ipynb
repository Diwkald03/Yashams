##9. Convert the tuple `("red", "green", "blue")` into a list, add `"yellow"`, and convert it back into a tuple.  


```python
import builtins
# Clear any shadowed 'tuple' variable from previous executions to ensure the built-in function is used.
if 'tuple' in globals() and globals()['tuple'] is not builtins.tuple:
    del globals()['tuple']

# Define the original tuple
colors_tuple = ("red", "green", "blue")
print(f"Original tuple: {colors_tuple}")
print(f"Type: {type(colors_tuple)}")

print("\n" + "-"*40 + "\n")

colors_list = list(colors_tuple)
print(f"Converted to list: {colors_list}")
print(f"Type: {type(colors_list)}")

print("\n" + "-"*40 + "\n")

colors_list.append("yellow")
print(f"After adding 'yellow': {colors_list}")

print("\n" + "-"*40 + "\n")

colors_tuple_new = tuple(colors_list)
print(f"Converted back to tuple: {colors_tuple_new}")
print(f"Type: {type(colors_tuple_new)}")

print("\n" + "-"*40 + "\n")

print("Summary:")
print(f"Original tuple:  {('red', 'green', 'blue')}")
print(f"Final tuple:     {colors_tuple_new}")
print(f"Elements added:  'yellow'")
```
